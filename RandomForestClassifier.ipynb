```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data.csv')
n = 8 # сдвиг и кол-во дополненных столбцов
for column in df.columns[2:17]:
  for i in range(n):
    df[f'{column}_back_{i}'] = df[f'{column}'].shift(i+1) # if df.index%128 != 0 else df[f'{n}']
df[['F3','F3_back_0']]
df.columns
```

```python
df_list = []
for i in range(100):
  df_list.append(f'df_{i}')
  df_list[i] = df.iloc[640*i+n:640*(i+1)]
df = pd.concat(df_list)
df
```

```python
df['forward'] = (df['class'] == 'forward').astype(int)
df['back'] = (df['class'] == 'back').astype(int)
df['right'] = (df['class'] == 'right').astype(int)
df['left'] = (df['class'] == 'left').astype(int)
df['neutral'] = (df['class'] == 'neutral').astype(int)
# df = df.drop(['class'], axis=1)
# df = df[['iter',
#          'forward',
#          'back',
#          'right',
#          'left',
#          'neutral',
#          'F3', 'FC5', 'AF3', 'F7', 'T7', 'P7', 'O1', 'O2', 'P8',
#        'T8', 'F8', 'AF4', 'FC6', 'F4', 'F3_back_0', 'F3_back_1', 'FC5_back_0',
#        'FC5_back_1', 'AF3_back_0', 'AF3_back_1', 'F7_back_0', 'F7_back_1',
#        'T7_back_0', 'T7_back_1', 'P7_back_0', 'P7_back_1', 'O1_back_0',
#        'O1_back_1', 'O2_back_0', 'O2_back_1', 'P8_back_0', 'P8_back_1',
#        'T8_back_0', 'T8_back_1', 'F8_back_0', 'F8_back_1', 'AF4_back_0',
#        'AF4_back_1', 'FC6_back_0', 'FC6_back_1', 'F4_back_0', 'F4_back_1']]
df
```

```python
forest = RandomForestClassifier()
x_start = 2
x_stop = y_start = 14*(n+1) + 2
y_stop = 14*(n+1) + 2 + 5
X = df.values[:, x_start:x_stop]
Y = df.values[:, y_start:y_stop].astype(int)
trainX, testX, trainY, testY = train_test_split( X, Y, test_size = 0.1)
model = forest.fit(trainX, trainY)
```

```python
df_1 = pd.DataFrame(testY*model.predict(testX))
(df_1.values == 1.0).sum()/df_1.shape[0]
```
