```python
import pandas as pd
```

```python
df= pd.read_csv("/content/2222.csv",sep = ';',header=None,encoding= 'cp1251')
df
```

```python
# Создадим результирующую таблицу 
result = df.drop([1, 2, 7], axis=1) # убираем ненужные столбцы (1, 2, 7)
result = result[result.index%5==0].reset_index(drop=True) # оставляем строки 0 и 5
result.columns = ['Номенклатура', 'Компания1', 'Цена1', 'Компания2', 'Цена2'] # меняем названия колонок
result # убедимся что всё правильно
```

```python
# Данный цикл сделан для случаев, если данных было бы не на 10 строчек, а на сколько угодно. Можно было, конечно, обойтись и без него

# Возьмём имена компаний, чтоб заполнить пропуски в итоговой таблице
for i in range (2):
  pre_df = df[5*i: 5*(i+1)].reset_index(drop=True) # разделим датасет по 5 строк (т.к. 5 строк на каждую позицию номенклатуры)
  pre_df = pre_df[pre_df.index % 2 == 1] # оставим строки с именами компаний
  result.iloc[i, 1], result.iloc[i, 3] = list(pre_df[0]) # взяли имена компаний и поместили в наш итоговый датасет
```
