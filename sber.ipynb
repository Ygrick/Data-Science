# Блок с формированием исходной витрины (в полном соответствии с таблицей 1)

```python
df = pd.read_excel('table1.xlsx')
df
```

# Блок, предназначенный для формирования целевой витрины 

```python
# заменили даты с 9999 годом на null
df['finish_date'] = pd.to_datetime(df['finish_date'], errors = 'coerce')
df
```

```python
# заменили null в finish_date на дату из столбца end_da
df['finish_date'] = df['finish_date'].fillna(df['end_da'])
df
```

```python
# удалили ненужный столбец
df = df.drop('end_da', axis=1)
df
```

```python
# все даты находятся в нужном диапазоне: 01.09.2020 – 31.12.2020
# в примере(табл 3) дата у 16234 начинается с 21.09.2020, а в табл1 с 20.09.2020
# у 17345 вообще один пример(в табл 3), хотя в табл1 он прописан два раза
# в связи с этим, первый пункт считаю выполненным правильно
```

```python
# Вычислим день недели для каждого дня (от 2 сентября до 31 декабря)

# Список дней
date_range_for_days_of_week = pd.date_range(
    '2020-09-02',
    '2020-12-31').strftime('%d/%m/%Y').tolist()
date_range_for_days_of_week

# список рабочий\выходной день
day_of_work = []
for i in range(len(date_range_for_days_of_week)):
  if i%7 in [3, 4]:
    day_of_work.append(0)
  else:
    day_of_work.append(1)

# создание сводной таблицы
df_days_of_week =  pd.DataFrame({'ymd_date': date_range_for_days_of_week,
                                 'day_of_work': day_of_work
                                 })
df_days_of_week
```

```python
# для типов [1:3] мы генерируем список. Если день недели рабочий (==1), 
# то в список добавляется значение, характеризующее работает ли сотрудник в офисе или нет
# 0 - дома, 
# 1 - в офисе, 
# Nan - неизвестно (соответствует типу 3 - смешанный режим работы)
def at_work_for_type_1_2_3(type_of_work, day_of_work):
  return list(map(lambda x: type_of_work if x == 1 else float('Nan'), 
                             day_of_work
                             ))
```

```python
# для типов [3:4] мы генерируем список. 
# Смотрим на каждый день недели. Если день недели выходной (== 0), 
# то считаем его (т.к. после двух\четырёх выходных(в зависимости от типа) мы должны сменить локацию - дом\офис)
# и добавляем в список Nan
# Если день недели рабочий (==1), то смотрим - прошло ли два\четыре выходных чтоб поменять график (и меняем, если требуется)

def at_work_for_type_3_4(type_work, day_of_work):
  # сколько нам считать выходных в зависимости от типа
  count_weekdays = type_work - 2 
  # счёт выходных
  n = 0
  # список в которые будем класть значения
  to_be_at_office = []
  # работаем ли мы на этой неделе в офисе (первоначально - нет)
  week_office = 0

  for item in day_of_work:
    # если день выходной
    if item == 0:
      # добавляем значение в список
      to_be_at_office.append(float('Nan'))
      n += 1
    # иначе
    else:
      # прошло ли два\четыре выходных
      if n == 2*count_weekdays:
        # меняем тип работы
        week_office = (week_office == 0) + 0
        n = 0
      # добавляем значение в список
      to_be_at_office.append(week_office)
  return to_be_at_office
```

```python
# будем создавать список датафреймов, которые в последствии объединим в один
list_df = []
for i, item in enumerate(df['tab_num']):
  # тип работы (0,1,2,3,4)
  type_of_work = df[i:i+1]['wplace_type'].values[0]

  # создаем список дат в которые работал сотрудник
  date_range = pd.date_range(df[i:i+1]['start_date'].values[0],
                             df[i:i+1]['finish_date'].values[0]
                             ).strftime('%d/%m/%Y').tolist()

  # создаём датафрейм табельный номер - дата работы
  temporary_df = pd.DataFrame({'tab_num': item,
                               'ymd_date': date_range,
                               })
  # соединяем его с датафреймом отвечающим за день недели (чтоб знать когда выходной\рабочий день)
  temporary_df = temporary_df.merge(df_days_of_week, 
                                    how='left', 
                                    on='ymd_date')

  # на основе типа работы сотрудника формируем график
  if type_of_work < 2:
    to_be_at_office = at_work_for_type_1_2_3(type_of_work, 
                                             temporary_df['day_of_work']
                                             ) 
  elif type_of_work == 2:
    to_be_at_office = at_work_for_type_1_2_3(float('Nan'), 
                                             temporary_df['day_of_work']
                                             )
  else:
    to_be_at_office = at_work_for_type_3_4(type_of_work, 
                                           temporary_df['day_of_work']
                                           ) 
  
  # check result
  print(to_be_at_office)
  # добавляем столбец с графиком в наш датафрейм
  temporary_df['to_be_at_office'] = to_be_at_office
  # удаляем вспомогательный столбец с днями недели
  temporary_df = temporary_df.drop('day_of_work', axis=1)
  # добавляем датафрейм в список датафреймов
  list_df.append(temporary_df)
```

```python
# check result
list_df
```

```python
# соединяем датафреймы отбрасывая индексы
df_output = pd.concat(list_df).reset_index(drop=True)
```

```python
# сохраняем датафрейм в эксель
df_output.to_excel("output.xlsx")
```
