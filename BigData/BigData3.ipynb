```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
import numpy as np
```

```python
X = data['target_names']
```

```python
Y = df['target'].value_counts()
```

```python
px.bar(x=X, y=Y,  text=Y)
```

```python
# 3.1
fig = go.Figure(px.bar(x=X, 
                       y=Y, 
                       text=Y
                       ))
# 3.2
fig.update_traces(marker=dict(color=Y, coloraxis="coloraxis"))
# 3.3
fig.update_traces(marker=dict(line=dict(color='black', width=2)))
# 3.4
fig.update_layout(title={'text': 'Targets_count', 
                         'font_size': 20, 
                         'y':0.95,
                         'x':0.5,
                         'xanchor': 'center',
                         'yanchor': 'top'})
# 3.5
fig.update_layout(xaxis_title= "Название класса", 
                  xaxis_title_font_size = 16, 
                  xaxis_tickangle=315,
                  yaxis_title="Количество значений класса",
                  yaxis_title_font_size=16
                  )
# 3.6
fig.update_layout(xaxis_tickfont_size=14,
                  yaxis_tickfont_size=14,
                  )
# 3.7
fig.update_layout(width=None,
                  height=700
                  )
# 3.8
fig.update_layout(margin=dict(l=0,r=0,t=0,b=0))

fig.show()
```

```python
# 4
figpie = go.Figure(px.pie(values=Y, names=X))
figpie.update_traces(marker=dict(colors=Y))
figpie.update_traces(marker=dict(line=dict(color='black', width=2)))
figpie.update_layout(title={'text': 'Targets_count', 
                         'font_size': 20, 
                         'y':0.95,
                         'x':0.5,
                         'xanchor': 'center',
                         'yanchor': 'top'},
                     xaxis_title= "Название класса", 
                     xaxis_title_font_size = 16, 
                     xaxis_tickangle=315,
                     yaxis_title="Количество значений класса",
                     yaxis_title_font_size=16,
                     xaxis_tickfont_size=14,
                     yaxis_tickfont_size=14,
                     width=None,
                     height=700,
                     margin=dict(l=0,r=0,t=100,b=0),
                     )
figpie.show()
```

```python
# 5
figscat = go.Figure(go.Scatter(x=df.index, 
                               y=df['alcohol'], 
                               mode='lines+markers'))
# 5.1
figscat.update_traces(line_color='crimson', 
                      marker_color='white', 
                      marker_line_color='black', 
                      marker_line_width=2)
# 5.2
figscat.update_xaxes(gridcolor='ivory', gridwidth=2)
figscat.update_yaxes(gridcolor='ivory', gridwidth=2)
figscat.update_layout(title={'text': 'Targets_count', 
                         'font_size': 20, 
                         'y':0.95,
                         'x':0.5,
                         'xanchor': 'center',
                         'yanchor': 'top'},
                     xaxis_title= "Величина алкоголя", 
                     xaxis_title_font_size = 16, 
                     xaxis_tickangle=315,
                     yaxis_title="Порядковый номер",
                     yaxis_title_font_size=16,
                     width=None,
                     height=700,
                     margin=dict(l=0,r=0,t=100,b=0),
                     )

figscat.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')     # Стиль графиков, можно закомментировать
plt.rcParams['figure.figsize'] = (10, 7.5)  
```

```python
# 6 
plt.bar(x=X, height=Y, color='teal', edgecolor='black', linewidth=5)
barlist=plt.bar(x=X, height=Y, color='teal', edgecolor='black')
barlist[0].set_color('r')
barlist[1].set_color('g')
barlist[2].set_color('b')
plt.title('Targets_bar')
plt.ylabel('counts')
plt.xlabel('n_class')
```

```python
plt.pie(Y, labels=X, autopct=str)
plt.title('Targets_pie')
```

```python
big=13
med=10
small=2
params = {'figure.figsize': (15,10), #размер внешних границ графика
          'axes.titlesize': big, #размер заголовка графика
          'axes.labelsize': med, #размер наименований осей
          'lines.linewidth': small, #толщина линии графика
          'lines.markersize': small*2, #толщина марекрков графика
          'xtick.labelsize': med, #размер меток оси абсцисс
          'ytick.labelsize': med, #размер меток оси ординат
          'lines.marker': '.',
          'lines.markerfacecolor': 'white', #основной цвет метки
          'lines.markersize': 10, #основной размер метки
          'lines.markeredgecolor': 'black', #цвет контура метки
          'lines.markeredgewidth': 1, #размер контура метки
          'axes.grid': True} # включение сетки
plt.rcParams.update(params)
plt.plot(df.index,
         df['alcohol'],
         )
plt.title('Targets_line')
```
