```python
import pandas as pd
```

```python
df = pd.read_csv('ECDCCases.csv')
df
```

```python
# Проверить в данных наличие пропущенных значений.
df.isna().sum()
```

```python
# Вывести количество пропущенных значений в процентах. 
df_of_nulls = pd.DataFrame({
    'name_column': [item for item in df.isna().keys()],
    'values of nulls %': [str(round(df.isna().sum()[item]*100/len(df),5)) + '%' for item in df.isna().keys()],
    })
df_of_nulls
```

```python
# Удалить два признака, в которых больше всех пропущенных значений
df.drop(
    df_of_nulls.sort_values('values of nulls %', ignore_index=True)['name_column'][-2:], 
    axis=1,
    inplace=True
    )
df
```

```python
# Для оставшихся признаков обработать пропуски
for item in df.keys():
  if df.dtypes[item] == object:
    df[item].fillna('other',
                    inplace=True)
  else:
    df[item].fillna(df[item].median(),
                    inplace=True)
```

```python
# Показать, что пропусков больше в данных нет.
df.isna().sum()
```

```python
# Посмотреть статистику по данным, используя describe(). 
# Сделать выводы о том, какие признаки содержат выбросы.
df.describe() # cases	deaths содержат выборсы
```

5. Загрузить данные из файла “ bmi.csv”. Взять оттуда две выборки. Одна
выборка – это индекс массы тела людей c региона northwest, вторая выборка –
это индекс массы тела людей с региона southwest. Сравнить средние значения
этих выборок, используя t-критерий Стьюдента. Предварительно проверить
выборки на нормальность (критерий Шопиро-Уилка) и на гомогенность
дисперсии (критерий Бартлетта).


```python
df_bmi = pd.read_csv('bmi.csv')
df_bmi
```

```python
df_bmi_northwest = df_bmi[df_bmi['region'] == 'northwest'].reset_index(drop=True)
df_bmi_northwest
```

```python
df_bmi_southwest = df_bmi[df_bmi['region'] == 'southwest'].reset_index(drop=True)
df_bmi_southwest
```

```python
import scipy.stats as sts
```

```python
res_northwest = sts.shapiro(df_bmi_northwest['bmi'])
res_southwest = sts.shapiro(df_bmi_southwest['bmi'])
print(res_northwest)
print(res_southwest)
# Тест дает p-значение выше 0.05, следовательно, выборки имеют нормальное распределение.
```

```python
sts.bartlett(df_bmi_northwest['bmi'], df_bmi_southwest['bmi'])
# p-уровень превышает 0.05, следовательно, дисперсии выборок схожи
```

```python
sts.ttest_ind(df_bmi_northwest['bmi'], df_bmi_southwest['bmi'])
# p-значение ниже 0.05, нулевая гипотеза отвергается, средние значения выборок различны
```

7. С помощью критерия Хи-квадрат проверить, являются ли переменные
зависимыми. 

  Использовать функцию scipy.stats.chi2_contingency(). 

  Влияет ли семейное положение на занятость?

```python
data = pd.DataFrame({'Женат': [89,17,11,43,22,1],
                     'Гражданский брак': [80,22,20,35,6,4],
                     'Не состоит в отношениях': [35,44,35,6,8,22]
                     })
data.index = ['Полный рабочий день',
              'Частичная занятость',
              'Временно не работает',
              'На домохозяйстве',
              'На пенсии',
              'Учёба'
              ]
data
```

```python
chi2, p_value, dof = sts.chi2_contingency(data)[:3]
print(f' статистика: {chi2} \n p-value: {p_value} \n степени свободы: {dof}')
# Влияет ли семейное положение на занятость? -> p-value < 0.05 -> да, влияет
```

```python
res1 = sts.shapiro(data.transpose()['Полный рабочий день'])
res2 = sts.shapiro(data.transpose()['Частичная занятость'])
print(res1)
print(res2)
# Тест дает p-значение выше 0.05, следовательно, выборки имеют нормальное распределение.
print(sts.bartlett(res1, res2))
# p-уровень превышает 0.05, следовательно, дисперсии выборок примерно одинаковы
print(sts.ttest_ind(res1, res2))
# p-значение намного выше 0.05, выборки схожи между собой -> семейное положение НЕ влияет на занятость
```
