```python
def task1():
  numbers = []
  number = int(input())
  numbers.append(number)

  while sum(numbers) != 0:
    number = int(input())
    numbers.append(number)

  return sum([num**2 for num in numbers])
```

```python
task1()
```

```python
def task2(n):
  numbers = []
  for i in range(n):
    for _ in range(i+1):
      numbers.append(i+1)
  return numbers[:n]
```

```python
print(*task2(12))
```

```python
def task3():
  matrix = [
      [1,1,1,2],
      [1,1,1,3],
      [1,1,1,4],
      [1,1,1,5]
  ]
  items = []
  for m in matrix:
    items += m
  return items
```

```python
task3()
```

```python
def task4(a, b):
  keys = list(set(b))
  keys.sort()
  result = {}
  for key in keys:
    result[key] = sum([a[i] 
                       for i, symbol in enumerate(b) 
                       if symbol==key]) 

  return result
```

```python
A = [1, 2, 3, 4, 2, 1, 3, 4, 5, 6, 5, 4, 3, 2]
B = ['a', 'b', 'c', 'c', 'c', 'b', 'a', 'c', 'a', 'a', 'b', 'c', 'b', 'a']
task4(A, B)
```
